<h2> Saat Uygulaması </h2>
* Dijital ve Analog yazısı bulunan kurbağa sembolleri ile saatler arasında geçiş yapılabilir. <br>
* Dijital formatta 24 saat ve 12 saat formatında gösterim seçeneği seçilebilir.<br>
* Alarm butonu ile istenilen zamana alarm kurulabilir. Alarm zamanı geldiğinde bir pop-up ekranı çıkacaktır.<br>

```python
# Analog Saat Uygulamasi
import turtle
import time
import datetime as dt
from datetime import date

analog = True # baslangicta analog saat cizimi
dijital = False
turtle.tracer(0) # bu satir sayesinde turtle bekleme olmadan tum cizimleri yapar
turtle.bgcolor("white") # beyaz arkaplan rengi
turtle.setup(600,600) # pencere boyutu
turtle.penup()

# Baslik Gosterimi
baslikPen = turtle.Turtle() # baslik kalemi
baslikPen.hideturtle()
baslikPen.penup()
baslikStil = ('Courier', 30, 'bold')
def baslik(str): # baslik yazma fonksiyonu
    baslikPen.clear()
    baslikPen.goto(0,250) 
    baslikPen.write(str, font=baslikStil, align='center')
baslik("Analog Saat")

# Tarih Gosterimi
tarihStil = ('Courier', 13, 'bold')
gun = dt.datetime.now().day
ay = dt.datetime.now().month
yil = dt.datetime.now().year
turtle.goto(50, -250) # tarih kutucugu cizimi
turtle.pendown()
turtle.setheading(270) 
turtle.forward(25)
turtle.right(90)
turtle.forward(100)
turtle.write(str(gun).zfill(2) + "/"+str(ay).zfill(2)+"/"+ str(yil).zfill(2), font=tarihStil) # tarih yazdırma komutu
turtle.right(90)
turtle.forward(25)
turtle.right(90)
turtle.forward(100)
turtle.penup()

turtle.goto(0,0)
turtle.hideturtle()
```

```python
# Alarm Butonu 
butonStil = ('Courier', 10, 'italic')
alarmPen = turtle.Turtle()
alarmPen.hideturtle()
alarmPen.shape("triangle") # kirmizi ucgen alarm butonu
alarmPen.fillcolor("red")
alarmPen.penup()
alarmPen.goto(-260, 180)
alarmPen.write("Alarm", align='center', font=butonStil)
alarmPen.sety(170)
alarmPen.showturtle()
alarmSa = None 
alarmDk = None 
alarmSn = None 

def alarmFun(x, y): # alarm butonuna basildiginda calisacak fonksiyon
    alarmZamani = turtle.textinput('Alarm', 'Alarm kurmak istediginiz zamani arasinda ":" kullanarak 24 saat formatinda "SAAT:DAKIKA:SANIYE" olacak sekilde giriniz: ')
    global alarmSa
    global alarmDk
    global alarmSn
    alarmSa, alarmDk, alarmSn = alarmZamani.split(":") # giris : ile bolunerek tek tek gerekli degiskenlere ataniyor
    alarmPen.sety(140)
    alarmPen.clear()
    alarmPen.write("Alarm Zamani: " + str(alarmSa).zfill(2) + ":"+str(alarmDk).zfill(2)+ ":" + str(alarmSn).zfill(2))
    alarmPen.sety(170)
```

```python
# Analog dijital degisimi
dijitalPen = turtle.Turtle() # dijital buton kalemi
dijitalPen.pensize(3)
dijitalSaatPen = turtle.Turtle() # dijital saat kalemi
dijitalSaatPen.hideturtle()
dijitalSaatPen.goto(-150,-75)
dijitalSaatPen.color("#03CE03")

sa24Pen = turtle.Turtle() # 24 saat gosterim butonu kalemi
sa12Pen = turtle.Turtle() # 12 saat gosterim butonu kalemi
sa24 = True # baslangicta 24 saat gosterimi olarak ayarli
sa12 = False

def dijitalFun(x, y): # dijital butonuna basildiginda calisacak fonksiyon
    global analog
    analog = False
    global dijital
    dijital = True
    
    # analog gosterimden kalanlari silme
    cercevePen.clear() 
    saatArasiPen.clear()
    numaraPen.clear()
    saatPen.clear()
    dakikaPen.clear()
    saniyePen.clear()
    merkezPen.clear()  
    
    # 24 ve 12 saat gosterim butonlarinin cizilmesi
    sa24Pen.hideturtle()
    sa24Pen.shape("circle")
    sa24Pen.fillcolor("blue")
    sa24Pen.penup()
    sa24Pen.goto(-75, 200)
    sa24Pen.write("24 Saat Formati", align='center', font=butonStil)
    sa24Pen.sety(180)
    sa24Pen.showturtle()
    
    sa12Pen.hideturtle()
    sa12Pen.shape("circle")
    sa12Pen.fillcolor("blue")
    sa12Pen.penup()
    sa12Pen.goto(75, 200)
    sa12Pen.write("12 Saat Formati", align='center', font=butonStil)
    sa12Pen.sety(180)
    sa12Pen.showturtle()
    
    # Dijital saat kutucugu cizimi
    dijitalPen.forward(150)
    dijitalPen.right(90)
    dijitalPen.forward(75)
    dijitalPen.right(90)
    dijitalPen.forward(300)
    dijitalPen.right(90)
    dijitalPen.forward(75)
    dijitalPen.right(90)
    dijitalPen.forward(150)
    dijitalPen.hideturtle()
    baslik("Dijital Saat")
    
def analogFun(x, y): # analog butonuna basildiginda calisacak fonksiyon
    global analog
    analog = True
    global dijital
    dijital = False
    
    baslik("Analog Saat") # basligi degistirme
    cerceveCizme()
    saatArasiCizme()
    rakamYazma()
    
def sa24Fun(x,y): # 24 saat gosterimi butonuna basildiginda calisacak fonksiyon
    global sa24
    global sa12
    sa24 = True
    sa12 = False
    
def sa12Fun(x,y): # 12 saat gosterimi butonuna basildiginda calisacak fonksiyon
    global sa24
    global sa12
    sa24 = False
    sa12 = True
    
# Dijital Butonu Cizimi
DbutonPen = turtle.Turtle()
DbutonPen.hideturtle()
DbutonPen.shape("turtle")
DbutonPen.fillcolor("black")
DbutonPen.penup()
DbutonPen.goto(-260, 240)
DbutonPen.write("Dijital", align='center', font=butonStil)
DbutonPen.sety(230)
DbutonPen.showturtle()

# Analog Butonu Cizimi
AbutonPen = turtle.Turtle()
AbutonPen.hideturtle()
AbutonPen.shape("turtle")
AbutonPen.fillcolor("black")
AbutonPen.penup()
AbutonPen.goto(-200, 240)
AbutonPen.write("Analog", align='center', font=butonStil)
AbutonPen.sety(230)
AbutonPen.showturtle()

dijitalStil = ('Courier', 45, 'italic')
```

```python
# Saat Çerçevesini Çizme
cercevePen = turtle.Turtle()
def cerceveCizme():
    cercevePen.speed(0)
    cercevePen.penup()
    cercevePen.pensize(3)
    cercevePen.goto(0,-190) # saat cemberini cizmek icin
    cercevePen.pendown()
    cercevePen.begin_fill()
    cercevePen.fillcolor("#DEDEDE") # gri saat arkaplan rengi
    cercevePen.circle(190)
    cercevePen.end_fill()
    cercevePen.penup()
    cercevePen.goto(0,0)
    cercevePen.hideturtle()
    for i in range(12): # Cizgileri cizen dongu
        cercevePen.forward(150)
        cercevePen.pendown()
        cercevePen.forward(40)
        cercevePen.penup()
        cercevePen.goto(0,0)
        cercevePen.right(30) # 30 derece saga dondurme
cerceveCizme()
```

```python
# Saat Arasini 5 Parcaya Bolme
saatArasiPen = turtle.Turtle()
def saatArasiCizme():
    saatArasiPen.hideturtle()
    saatArasiPen.penup()
    saatArasiPen.speed(0)
    saatArasiPen.pensize(1)
    saatArasiPen.setheading(90)
    for i in range(60): # dakika cizgilerini cizme
        saatArasiPen.forward(160)
        saatArasiPen.pendown()
        saatArasiPen.forward(20)
        saatArasiPen.penup()
        saatArasiPen.goto(0,0)
        saatArasiPen.right(6)     
saatArasiCizme()
```

```python
# Saat Rakamlarını Yazma
numaraPen = turtle.Turtle()
def rakamYazma():
    numaraPen.hideturtle()
    numaraPen.penup()
    numaraPen.speed(0)
    numaraPen.pencolor("black")
    style = ('Courier', 25, 'italic')
    numaraPen.goto(0,-20)
    numaraPen.setheading(60) # ilk rakam olan 1 i yazmak icin turtle yi dondurduk
    for i in range(1,13,1): # saat rakamlarini yazma
        numaraPen.forward(220)
        numaraPen.pendown()
        numaraPen.write(i, font=style, align='center')
        numaraPen.penup()
        numaraPen.goto(0,-20)
        numaraPen.right(30)
rakamYazma()
```

```python
# Saat, Dakika ve Saniye İbresi Çizme
def SaatMerkeziCizme(): # akrep, yelkovan ve saniye cizgilerinin birlestigi yere bir daire cizdik
    merkezPen.goto(0, -10)
    merkezPen.pencolor("gray")
    merkezPen.fillcolor("black")
    merkezPen.pensize(2)
    merkezPen.begin_fill()
    merkezPen.circle(10)
    merkezPen.end_fill()
    
def AkrepIbreCizme(aci):
    saatPen.clear()
    saatPen.penup()
    saatPen.goto(0,0)
    saatPen.setheading(90) # turtle 12 yi gosteriyor
    saatPen.right(aci) # turtle gosterilmesi gereken saate donuyor
    saatPen.speed(0)
    saatPen.pencolor("green")
    saatPen.fillcolor("green")
    saatPen.begin_fill()
    saatPen.pendown()
    saatPen.forward(79.89) # once ikizkenar ucgenin yuksekligi ciziliyor
    saatPen.right(183) # gerekli aci kadar donduruluyor
    saatPen.forward(80) # ikizkenarin ilk kenari
    saatPen.right(267) # gerekli aci kadar donduruluyor
    saatPen.forward(8.37) # ikizkenarin tabani
    saatPen.right(267) # gerekli aci kadar donduruluyor
    saatPen.forward(80) # ikizkenarin ikinci kenari
    saatPen.end_fill()
    
def YelkovanIbreCizme(aci):
    dakikaPen.clear()
    dakikaPen.penup()
    dakikaPen.goto(0,0)
    dakikaPen.setheading(90) # turtle 12 yi gosteriyor
    dakikaPen.right(aci) # turtle gosterilmesi gereken saate donuyor
    dakikaPen.speed(0)
    dakikaPen.pencolor("blue")
    dakikaPen.fillcolor("blue")
    dakikaPen.begin_fill()
    dakikaPen.pendown()
    dakikaPen.forward(124.98) # once ikizkenar ucgenin yuksekligi ciziliyor
    dakikaPen.right(181) # gerekli aci kadar donduruluyor
    dakikaPen.forward(125) # ikizkenarin ilk kenari
    dakikaPen.right(269) # gerekli aci kadar donduruluyor
    dakikaPen.forward(4.36) # ikizkenarin tabani
    dakikaPen.right(269) # gerekli aci kadar donduruluyor
    dakikaPen.forward(125) # ikizkenarin ikinci kenari
    dakikaPen.end_fill()
    
def SaniyeIbreCizme(aci):
    saniyePen.clear()
    saniyePen.penup()
    saniyePen.goto(0,0)
    saniyePen.setheading(90) # turtle 12 yi gosteriyor
    saniyePen.right(aci) # turtle gosterilmesi gereken saate donuyor
    saniyePen.speed(0)
    saniyePen.pencolor("red")
    saniyePen.fillcolor("red")
    saniyePen.begin_fill()
    saniyePen.pendown()
    saniyePen.forward(149.98) # once ikizkenar ucgenin yuksekligi ciziliyor
    saniyePen.right(181) # gerekli aci kadar donduruluyor
    saniyePen.forward(150) # ikizkenarin ilk kenari
    saniyePen.right(269) # gerekli aci kadar donduruluyor
    saniyePen.forward(5.24) # ikizkenarin tabani
    saniyePen.right(269) # gerekli aci kadar donduruluyor
    saniyePen.forward(150) # ikizkenarin ikinci kenari
    saniyePen.end_fill()
    SaatMerkeziCizme() # saat merkezi dairesinin en ustte olmasi icin her saniye ciziminden sonra tekrardan cagiriyoruz

#ibrelerin kalemleri
saatPen = turtle.Turtle()
dakikaPen = turtle.Turtle()
saniyePen = turtle.Turtle()
merkezPen = turtle.Turtle()
saatPen.hideturtle()
dakikaPen.hideturtle()
saniyePen.hideturtle()
merkezPen.hideturtle()
```

```python
# Butonlarin ustune basildiginda calismasini saglayan fonksiyonlar
DbutonPen.onclick(dijitalFun, 1)

AbutonPen.onclick(analogFun, 1)

alarmPen.onclick(alarmFun, 1)

sa24Pen.onclick(sa24Fun, 1)

sa12Pen.onclick(sa12Fun, 1)
```

```python
# Sonsuz Dongu
while(True):
    sn = dt.datetime.now().second # su anki saat degerlerini alma
    dk = dt.datetime.now().minute
    sa = dt.datetime.now().hour
    
    # Eger alarm zamani su anki saate esitse pop-up ekrani ortaya cikarir
    if str(sa) == str(alarmSa) and str(dk) == str(alarmDk) and str(sn) == str(alarmSn): 
        turtle.TK.messagebox.showinfo(title="Alarm!", message="Alarm!")
        
    # Eger analog gosterim modu seciliyse saat cizimi yapar
    if analog:
        aciSn = sn * 6 # saniye acisini ayarlama
        aciDk = dk * 6 + aciSn / 60 # yelkovan acisini ayarlama
        aciSa = sa * 30 +  aciDk / 12 # akrep acisini ayarlama
        sa24Pen.clear() # 24 ve 12 saat gosterim modlarini temizleme ve saklama
        sa12Pen.clear()
        sa24Pen.hideturtle()
        sa12Pen.hideturtle()
        AkrepIbreCizme(aciSa) # ibrelerin cizimi
        YelkovanIbreCizme(aciDk)
        SaniyeIbreCizme(aciSn)
        turtle.update() #tracer fonkisyonundan sonra guncelleme yapma
    elif dijital: # dijital mod seciliyse dijital saat gosterimi yapar
        dijitalSaatPen.clear()
        if sa24: # 24 saat gosterim formatı
            dijitalSaatPen.write(str(sa).zfill(2) + ":"+str(dk).zfill(2)+ ":" + str(sn).zfill(2), font=dijitalStil) 
        elif sa12: # 12 saat gosterim formati
            if sa >= 0 and sa < 12:
                dijitalSaatPen.write(str(sa).zfill(2) + ":"+str(dk).zfill(2)+ ":" + str(sn).zfill(2) + " AM", font=dijitalStil)
            elif sa >= 12 and sa < 24:
                if sa >= 13:
                    saTemp = sa - 12
                dijitalSaatPen.write(str(saTemp).zfill(2) + ":"+str(dk).zfill(2)+ ":" + str(sn).zfill(2) + " PM", font=dijitalStil)
        turtle.update() #tracer fonkisyonundan sonra guncelleme yapma
        
turtle.done()
```
